```python
def giris():

  username = input("Kullanıcı Adınız: ")  #Kullanıcı Adı İnput
  passwd = input("Şifreniz: ")  #Şifre İnput

  with open("kullanici_bilgileri.txt", "r") as f:   # f olarak read olarak dosyayı aç
    sp = f.readlines()                              # readlines ile tüm satırları oku
    totl = [line.rstrip('\n') for line in sp]       # \n leri kaldırıyoruz

  suc = False                                       # for döngüsü sonrası hata döndürmek için boolen tanımlıyoruz.
  for i, user in enumerate(totl):                   # Enumerate ile indexliyoruz ve for döngüsüne giriyoruz.
    if (user.split(",")[0] == username) and (user.split(",")[1] == passwd):    # split ile , e göre ikiye ayırıp inputlardaki verilerlere karşılaştırma
      suc = True                                    # Hata döndürmemesi için True olarak değiştiriyoruz
      print("Giriş Yapıldı") 
  
  if suc == False:                                  # giriş yapılıp yapılmadığını kontrol ediyoruz.
    print("Kullanıcı Adınız Veya Şifreniz Hatalı!") #koşul sağlanırsa hatayı döndürüyoruz
  
  return suc

giris()
```
